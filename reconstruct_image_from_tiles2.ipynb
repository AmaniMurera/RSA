```python
city = src_path.split("/")[-3]
code = src_path.split("/")[-2]

def get_tile_name_path(dst_dir, index, city, code):
    '''
    Generating index-specific tile name and path
    '''
    dst_tile_name = f"{city}_{code}_{str(index).zfill(5)}.tif"
    dst_tile_path = os.path.join(dst_dir, dst_tile_name)
    return dst_tile_name, dst_tile_path

def merge_tiles(src_dir, dst_path):
    '''
    Merges tiles into a single georeferenced RGB image
    '''
    # Get the profile of the first tile
    tile_path = get_tile_name_path(src_dir, 1, city, code)[1]
    with rasterio.open(tile_path) as tile:
        meta = tile.meta.copy()
        # Check if the tile contains RGB bands
        if meta['count'] != 3:
            raise ValueError('The tile does not contain RGB bands')

    # Update the meta profile for the reconstructed image
    meta.update(width=data.width, height=data.height, count=3, dtype='uint8')

    with rasterio.open(dst_path, 'w', **meta) as dst:
        # Merge tile by tile
        for i in range(1, len(os.listdir(src_dir)) + 1):
            tile_path = get_tile_name_path(src_dir, i, city, code)[1]
            with rasterio.open(tile_path) as tile:
                # Calculate the window for the tile
                start_row = (i - 1) * tile.meta['height']
                end_row = min(start_row + tile.meta['height'], data.height)
                start_col = (i - 1) * tile.meta['width']
                end_col = min(start_col + tile.meta['width'], data.width)
                window = ((start_row, end_row), (start_col, end_col))

                # Read tile data and write to the destination
                tile_data = tile.read(window=window)
                dst.write(tile_data, window=window)

dst_reconstruct_path = os.path.join(dst_image_dir, 'reconstructed.tif')
merge_tiles(dst_image_dir, dst_reconstruct_path)
```
